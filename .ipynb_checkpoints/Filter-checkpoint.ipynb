```python
import pandas as pandaExport
```

```python
provinces = [('GUAYAS','GUAYAQUIL'),('AZUAY','CUENCA'),('PICHINCHA','QUITO')]
```

```python
for province,city in provinces:
    file = province+'.txt'
    records = pandaExport.read_csv(file,sep="	",encoding ="ISO-8859-1",error_bad_lines=False)
    print(province+'/'+city+' '+str(len(records)))
    city_filter = records['DESCRIPCION_CANTON'] == city
    records[city_filter].to_csv(city+'.csv')
```

```python
cuenca = 'CUENCA.csv'
pandaExport.read_csv(cuenca,sep=",",encoding ="ISO-8859-1",error_bad_lines=False)
```

#### Código para importar csv con información de los establecimientos de X ciudad

```python
import csv
import pandas as pandaExport
ciudad_x = pandaExport.read_csv(FILE_NAME,sep=",",encoding ="utf-8",error_bad_lines=False)
ciudad_x
```

```python
### obtenemos la cantidad de establecimientos por contribuyente
data = ciudad_x[["NUMERO_RUC","ESTADO_ESTABLECIMIENTO"]].groupby(["NUMERO_RUC"]).agg(['count']).reset_index()
data
```

```python
# hacemos un merge del dataframe con la cantidad de establecimiento y el dataframe inicial
df_merge = pandaExport.merge(left=ciudad_x,right=data, left_on='NUMERO_RUC', right_on='NUMERO_RUC',suffixes=("",""))
df_merge
```

```python
### agregamos un campo que permita conocer si ese establecimiento pertenece a un contribuyente con 3 o más locales
def objetivo(row):
    if row[("ESTADO_ESTABLECIMIENTO", "count")] >= 3:
        row["objetivo"] = True
    else:
        row["objetivo"] = False
    return row
df_data = df_merge.apply(objetivo, axis=1)
df_data
```

```python
### solo dejamos los contruyentes con una cantidad mayor o igual a dos establecimientos
df_objetivo = df_data[df_data["objetivo"]==True]
df_objetivo
```

```python
## guardar aquellos en un archivo para su posterior analisis.
export_csv = df_objetivo.to_csv("finalMayorIgual3_"+ FILE_NAME, index = False, header=True,encoding="utf-8")
```

```python
data_array =[]
def filter_sucess(finalResult, originData):
    global data_array
    write = open(finalResult,"a")
    read = open(originData, "r")
    for x in read:
        data = x.split(",")
        id = data[3]
        if id not in data_array:
            print(id)
            data_array.append(id)
            data_write = str( data[0]) +"," + str( data[1]) +"," + str( data[2]) 
            write.write(data_write)
    read.close()
    write.close()

CITY="GUAYAQUIL" 

filter_sucess("U1Location"+ CITY +".csv", "ULocation"+ CITY +".csv")
```

```python
df = pd.read_csv('reembolsos_senadores.csv',sep=",")
import numpy as np
## Aquì se obtienen los valores de media y desviación estandar por refund_description
data = df[["refund_description","refund_value"]].groupby(["refund_description"]).agg(['mean','std']).reset_index()
# Aqui se aplica un merge para que todas las filas tengan la información de media y desviación estandar y optimizar la obtención del z-score
df_merge = pd.merge(left=df,right=data, left_on='refund_description', right_on='refund_description',suffixes=("",""))
df_merge
```

```python
def outlier(row):
    value_refund = row["refund_value"]
    z_score = (value_refund - row[("refund_value","mean")])/row[("refund_value","std")]
    if not np.abs(value_refund-row[("refund_value","mean")]) <= (3*row[("refund_value","std")]):
        row["outlier"] = True
    else:
        row["outlier"] = False
    return row
#aquí obtener el z-score. En la función outlier, a partir de z-score obtenido se verifica si el valor es un outlier o no
df_data = df_merge.apply(outlier, axis=1)
df_data
```

```python
# dejamos solo las columnas de interes
df_data = df_data[["refund_description","outlier"]]
# obtenemos la cantidad de outlier por refund_description
df_data[df_data["outlier"]==True].groupby(["refund_description"]).count()
```

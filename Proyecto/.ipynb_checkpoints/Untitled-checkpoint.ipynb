```python
import csv
import pandas as pandaExport
FILE_NAME = 'block_0.csv'
dataset = pandaExport.read_csv(FILE_NAME,sep=",",encoding ="utf-8")
dataset
```

```python
dataset["date_measure"] = pandaExport.to_datetime(dataset["date_measure"])
dataset["date_measure"]
```

```python
dataset["date_measure"].dt.year
```

```python
data = dataset["energy_kwh"] = pandaExport.to_numeric(dataset["energy_kwh"],downcast='float',errors='coerce')
```

```python
dataset[dataset["energy_kwh"].isnull()].count()
```

```python
dataset = dataset.dropna()
dataset
```

```python
data = dataset.loc[:,["energy_kwh"]]
```

```python
data = data.set_index(dataset["date_measure"])
```

```python
data.head()
```

```python
import seaborn as sns
data.plot()
```

```python
weekly = data.resample('W').sum()
%matplotlib
weekly.plot(figsize=(10,8),style=[':'])
```

```python
daily = data.resample('D').sum()
daily.rolling(30,center=True).sum().plot()
```

```python
daily..sort_values(['energy_kwh'], ascending=[False])
```

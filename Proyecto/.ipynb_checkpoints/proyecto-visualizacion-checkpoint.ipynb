#### Código para importar antigua estructura del csv con información de los establecimientos de Guayas

```python
import csv
import pandas as pandaExport
#inicialmente se encontraban separados en dos archivos de texto. Uno con los establecimientos activos y otros con los inactivos
FILE_NAME_ACTIVOS = 'datos/GUAYAS/GUAYAS ACTIVOS.txt'
FILE_NAME_INACTIVOS = 'datos/GUAYAS/GUAYAS NO ACTIVOS.txt'
inactivos =pandaExport.read_csv(FILE_NAME_INACTIVOS,sep="	",encoding ="ISO-8859-1",error_bad_lines=False)
activos =pandaExport.read_csv(FILE_NAME_ACTIVOS,sep="	",encoding ="ISO-8859-1")
```

```python
# creamos un solo dataframe con los establecimientos activos y no activos
todos = pandaExport.concat([activos, inactivos])
todos
```

```python
### obtenemos la cantidad de establecimientos por contribuyente
data = todos[["Numero Ruc","Estado Establecimiento"]].groupby(["Numero Ruc"]).agg(['count']).reset_index()
data
```

```python
# hacemos un merge del dataframe con la cantidad de establecimiento y el dataframe inicial
df_merge = pandaExport.merge(left=todos,right=data, left_on='Numero Ruc', right_on='Numero Ruc',suffixes=("",""))
```

```python
df_merge
```

```python
### agregamos un campo que permita conocer si ese establecimiento pertenece a un contribuyente con dos o más locales
def objetivo(row):
    if row[("Estado Establecimiento", "count")] >= 2:
        row["objetivo"] = True
    else:
        row["objetivo"] = False
    return row
df_data = df_merge.apply(objetivo, axis=1)
df_data
```

```python
### solo dejamos los contruyentes con una cantidad mayor o igual a dos establecimientos
df_objetivo = df_data[df_data["objetivo"]==True]
df_objetivo
```

```python
## guardar aquellos en un archivo para su posterior analisis.
export_csv = df_objetivo.to_csv("objetivoMayorIgual2.csv", index = True, header=True)
```

```python
import csv
import pandas as pandaExport
FILE_NAME = 'GUAYAQUIL.csv'
guayaquil = pandaExport.read_csv(FILE_NAME,sep=",",encoding ="utf-8",error_bad_lines=False)
guayaquil
```

```python
### obtenemos la cantidad de establecimientos por contribuyente
data = guayaquil[["NUMERO_RUC","ESTADO_ESTABLECIMIENTO"]].groupby(["NUMERO_RUC"]).agg(['count']).reset_index()
data
```

```python
# hacemos un merge del dataframe con la cantidad de establecimiento y el dataframe inicial
df_merge = pandaExport.merge(left=guayaquil,right=data, left_on='NUMERO_RUC', right_on='NUMERO_RUC',suffixes=("",""))
df_merge
```

```python
### agregamos un campo que permita conocer si ese establecimiento pertenece a un contribuyente con dos o más locales
def objetivo(row):
    if row[("ESTADO_ESTABLECIMIENTO", "count")] >= 2:
        row["objetivo"] = True
    else:
        row["objetivo"] = False
    return row
df_data = df_merge.apply(objetivo, axis=1)
df_data
```

```python
### solo dejamos los contruyentes con una cantidad mayor o igual a dos establecimientos
df_objetivo = df_data[df_data["objetivo"]==True]
df_objetivo
export_csv = df_objetivo.to_csv("finalMayorIgual2.csv", index = True, header=True,encoding="utf-8")
```

```python
# ACTUALIZAR GUAYAQUIL
import csv
import pandas as pandaExport
FILE_NAME = 'finalMayorIgual2.csv'
guayaquil = pandaExport.read_csv(FILE_NAME,sep=",",encoding ="utf-8",error_bad_lines=False)
guayaquil_lat = pandaExport.read_csv("resultadofinal.csv",sep=",",encoding ="utf-8",error_bad_lines=False)
guayaquil_lat
```

```python
guayaquil
```

```python
#### para utilizar
df_w_ll = pandaExport.merge(left=guayaquil,right=guayaquil_lat, left_on='ID', right_on='ID',suffixes=("",""), how="inner")
df_w_ll.to_csv("actualizado.csv",header=True,encoding="utf-8",index=False)
def validar_lat_log(row):
    if not row["LAT"].isdigit():
        row["validoLL"] = False
        return row
    if  -2.287573 <= float(row["LAT"]) <= -2.053362 and -79.95912 <= float(row["LOG"]) <= -79.856351:
        row["validoLL"] = True
    else:
        row["validoLL"] = False
    return row
    
df_w_ll = df_w_ll.apply(validar_lat_log, axis=1)
df_w_ll[df_w_ll["validoLL"] == True].to_csv("actualizadoVValidado.csv",header=True,encoding="utf-8",index=False)
```

```python
faltantes = pandaExport.merge(left=guayaquil,right=guayaquil_lat, left_on='ID', right_on='ID',suffixes=("",""), how="left")
faltantes
```

```python
faltantes_ = faltantes[faltantes["LAT"].isnull()]
faltantes_ = faltantes_.drop(['LAT',"LOG"], 1)
faltantes_.to_csv("faltante.csv",header=True,encoding="utf-8",index=False)
faltantes
```

```python
import csv
import pandas as pandaExport
FILE_NAME = 'actualizadoVValidado.csv'
valido = pandaExport.read_csv(FILE_NAME,sep=",",encoding ="utf-8",error_bad_lines=False)
valido = valido.sort_values(by=["NUMERO_RUC","FECHA_INICIO_ACTIVIDADES","FECHA_REINICIO_ACTIVIDADES"])
#data = guayaquil[["NUMERO_RUC","ESTADO_ESTABLECIMIENTO"]].groupby(["NUMERO_RUC"]).agg(['count']).reset_index()
valido[["NUMERO_RUC","FECHA_INICIO_ACTIVIDADES","FECHA_REINICIO_ACTIVIDADES"]]
```

```python
prev_ruc = None
prev_id = None
def validar(row):
    global prev_ruc, prev_id
    if prev_ruc != int(row["NUMERO_RUC"]):
        prev_ruc = row["NUMERO_RUC"]
        prev_id = int(row["ID"])
    else:
        if prev_ruc:
            row["destino"] = int(prev_id)
            row["peso"] =  1
            return row
    
valido = valido.apply(validar, axis=1)
```

```python
valido = valido[valido["destino"].notnull()]
valido
```

```python
valido["destino"] = valido["destino"].astype(int)
valido[["ID","LAT","LOG","destino","peso"]].to_csv("edge.csv",header=True,encoding="utf-8",index=False)
```

#### Código final para importar csv con información de los establecimientos de X ciudad

```python
import csv
import pandas as pandaExport
FILE_NAME = CIUDADES[1]
ciudad_x = pandaExport.read_csv(FILE_NAME,sep=",",encoding ="utf-8",error_bad_lines=False)
ciudad_x
```

```python
### obtenemos la cantidad de establecimientos por contribuyente
data = ciudad_x[["NUMERO_RUC","ESTADO_ESTABLECIMIENTO"]].groupby(["NUMERO_RUC"]).agg(['count']).reset_index()
data
```

```python
# hacemos un merge del dataframe con la cantidad de establecimiento y el dataframe inicial
df_merge = pandaExport.merge(left=ciudad_x,right=data, left_on='NUMERO_RUC', right_on='NUMERO_RUC',suffixes=("",""))
df_merge
```

```python
### agregamos un campo que permita conocer si ese establecimiento pertenece a un contribuyente con 3 o más locales
def objetivo(row):
    if row[("ESTADO_ESTABLECIMIENTO", "count")] >= 3:
        row["objetivo"] = True
    else:
        row["objetivo"] = False
    return row
df_data = df_merge.apply(objetivo, axis=1)
df_data
```

```python
### solo dejamos los contruyentes con una cantidad mayor o igual a dos establecimientos
df_objetivo = df_data[df_data["objetivo"]==True]
df_objetivo
```

```python
## guardar aquellos en un archivo para su posterior analisis.
export_csv = df_objetivo.to_csv("finalMayorIgual3_"+ FILE_NAME, index = False, header=True,encoding="utf-8")
```

## para normalizar los datos

```python
FILE_NAME = CIUDADES[0]
ciudad_norm = pandaExport.read_csv(FILE_NAME,sep=",",encoding ="utf-8",error_bad_lines=False)
ciudad_norm
```

```python
locales_x_parroquias = ciudad_norm[["DESCRIPCION_PARROQUIA","ID"]].groupby(["DESCRIPCION_PARROQUIA"]).count().reset_index()
locales_x_parroquias
```

```python
poblacion_x = pandaExport.read_csv("Poblacion" + FILE_NAME,sep=",",encoding ="utf-8",error_bad_lines=False)
poblacion_x
```

```python
factor = pandaExport.merge(left=poblacion_x,right=locales_x_parroquias, left_on='PARROQUIA', right_on='DESCRIPCION_PARROQUIA',suffixes=("",""))
factor
```

```python
factor["factor"] = factor["ID"]/factor["POBLACION"]
factor[["DESCRIPCION_PARROQUIA","POBLACION","AREA","factor","ID"]].to_csv("factorNormalizacion" + FILE_NAME,header=True,encoding="utf-8",index=False)
factor[["DESCRIPCION_PARROQUIA","POBLACION","AREA","factor","ID"]]
```
